```python
def longest_palindrome(s):
    n = len(s)
    if n == 0:
        return ""

    # Create DP table
    dp = [[False] * n for _ in range(n)]

    start = 0
    maxLength = 1

    # Every single character is a palindrome
    for i in range(n):
        dp[i][i] = True

    # Check for substrings of length 2
    for i in range(n - 1):
        if s[i] == s[i + 1]:
            dp[i][i + 1] = True
            start = i
            maxLength = 2

    # Check for substrings of length 3 or more
    for length in range(3, n + 1):  # length is the substring length
        for i in range(n - length + 1):
            j = i + length - 1  # Ending index of substring
            if s[i] == s[j] and dp[i + 1][j - 1]:
                dp[i][j] = True
                start = i
                maxLength = length

    return start, maxLength


# temp = input().strip().split()

# w = int(temp[1])
# h = int(temp[0])

h = 4
w = 6

"""
4 6
BANANA
BERGEN
CANNOT
FELLOW
"""

# strings = []
# for i in range(h):
#     strings.append(input().strip())

strings = ["BANANA", "BERGEN", "CANNOT", "FELLOW"]

strings_vertical = []
for i in range(w):
    strings_vertical.append("".join([strings[j][i] for j in range(h)]))


max_x = 0
loc_x_horizontal = 0
loc_y_horizontal = 0

max_y = 0
loc_x_vertical = 0
loc_y_vertical = 0
for i in range(h):
    loc, maxLength = longest_palindrome(strings[i])
    if maxLength > max_x:
        max_x = maxLength
        loc_x_horizontal = loc
        loc_y_horizontal = i

# print(loc_x_horizontal, loc_y_horizontal, max_x-1, loc_y_horizontal)

for i in range(w):
    loc, maxLength = longest_palindrome(strings_vertical[i])
    if maxLength > max_y:
        max_y = maxLength
        loc_x_vertical = i
        loc_y_vertical = loc

# print(loc_x_vertical, loc_y_vertical, max_y, loc_x_vertical)

"""
0 2 4 2  (0,2) to (4,2)
2 1 3 2  (2,1) to (3,2)


-> area = (0, 1) to (4, 3)
"""
x1 = min(loc_x_horizontal, loc_x_vertical)
y1 = min(loc_y_horizontal, loc_y_vertical)
x2 = max(loc_x_horizontal + max_x - 1, loc_x_vertical)
y2 = max(loc_y_horizontal, loc_y_vertical + max_y - 1)

area = (x2 - x1 + 1) * (y2 - y1 + 1)
print(area)
```
